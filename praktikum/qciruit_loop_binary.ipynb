```python
import itertools
from myutils import Preprocessing 
from myutils import Datasets as ds
from myutils import Helpers 
from myutils import Complexity_Measures as cm
from myutils import QuatumCircuit
import numpy as np
import os
import pandas as pd
import tensorflow as tf
import random
import json
from matplotlib import pyplot as plt
import seaborn as sns

#Magic Command, so changes in myutils module are reloaded
%load_ext autoreload
%autoreload 1
%aimport myutils
```

```python
n_training = 100
n_test = 1000
```

```python
data = ds.get_preprocessed_datasets("Autoencoder_binary")["Autoencoder_binary"]
for subset in data:

    np.random.seed(1337)

    print(subset)

    data_subset = data[subset]
    
    #scale data 0, 2*np
    x_train = Preprocessing.minmax_scaler(data_subset["x_train"] , min = 0,max = 2 * np.pi)
    x_test = Preprocessing.minmax_scaler(data_subset["x_test"], min = 0,max = 2 * np.pi)

    #convert label from 0/1 to [0,1]/[1,0]
    y_train_quantum = Preprocessing.convertlabel_to_quantumstates(data_subset["y_train_binary"])
    y_test_quantum = Preprocessing.convertlabel_to_quantumstates(data_subset["y_test_binary"])


    # sampling and splitting 
    random.seed(1337)

    x_train_subset, y_train_quantum_subset = Preprocessing.select_nExamples(x_train, y_train_quantum, n=n_training)
    x_test_subset, y_test_quantum_subset = Preprocessing.select_nExamples(x_test, y_test_quantum, n=n_test)

    
    qcircuit = QuatumCircuit()
    hist,params = qcircuit.train(
        x_train=x_train_subset,
        y_train=y_train_quantum_subset,
        x_test=x_test_subset,
        y_test=y_test_quantum_subset,
        epochs=10,
        verbose=1)


    #save the hist

    Helpers.store(hist, "measure/qcircuit_loop/Autoencoder_binary/{}" .format(subset),"hist")
   

```
